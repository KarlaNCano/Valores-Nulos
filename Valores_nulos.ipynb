```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Ventas_totales.csv')
data.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

**Métodos de Sustitución de Valores Nulos**

```python
#Realizamos una copia del dataframe
data1=data.copy()
```

```python
#Se sustituyeron los valores de la columna "otros" con el promedio o media de esta
data1["otros"]=data1["otros"].fillna(round(data["otros"].mean(),1))
```

```python
#Se sustituyeron los valores nulos de la columna "electronicos_articulos_hogar" con la media
data1["electronicos_articulos_hogar"]=data1["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
```

```python
#Se sustituyeron los valores nulos de la columna "indumentaria_calzado_textiles_hogar" por el valor numérico "10"
data1["indumentaria_calzado_textiles_hogar"]= data1["indumentaria_calzado_textiles_hogar"].fillna(10)
```

```python
#Se sustituyeron los valores nulos de la columna "salon_ventas" por el string "27591730.13 "
data1["salon_ventas"] =data1["salon_ventas"].fillna("27591730.13")
```

```python
#Se sustituyeron los valores nulos de la columna "tarjetas_debito" por valores no nulos hacia adelante con "forward fill"
data1["tarjetas_debito"] =data1["tarjetas_debito"].fillna(method="ffill")
```

```python
#Se sustituyeron los valores nulos de la columna "tarjetas_credito" por valores no nulos hacia atrás con backward fill"
data1["tarjetas_credito"] =data1["tarjetas_credito"].fillna(method="bfill")
```

```python
#Se sustituyeron los valores nulos de la columna "otros_medios" con el promedio o la media
data1["otros_medios"]=data1["otros_medios"].fillna(round(data1["otros_medios"].mean(),1))
```

```python
#Se sustituyeron los valores nulos de la columna "subtotal_ventas_alimentos_bebidas" con la mediana
data1["subtotal_ventas_alimentos_bebidas"]=data1["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].median(),1))
```

```python
#Se sustituyeron los valores nulos de la columna "bebidas" por el valor numérico "9"
data1["bebidas"]= data1["bebidas"].fillna(9)
```

```python
#Se sustituyeron los valores nulos de la columna "alamacen" por valores no nulos hacia adelante con el filtor por columna "forward fill"
data1["almacen"] =data1["almacen"].fillna(method="ffill")
```

```python
#Se sustituyeron los valores nulos de la columna "panaderia" por valores no nulos hacia atrás con el filtro por columna "backward fill"
data1["panaderia"] =data1["panaderia"].fillna(method="bfill")
```

```python
#Se sustituyeron los valores nulos de la columna "lacteos" por el valor numérico "5"
data1["lacteos"]= data1["lacteos"].fillna(5)
```

```python
#Se sustituyeron los valores nulos de la columna "carnes" por valores no nulos hacia adelante con el filtro por columna "forward fill"
data1["carnes"] =data1["carnes"].fillna(method="ffill")
```

```python
#Se sustituyeron los valores nulos de la columna "verduleria_fruteria" por valores no nulos hacia atrás con el filtro por columna "backward fill"
data1["verduleria_fruteria"] =data1["verduleria_fruteria"].fillna(method="bfill")
```

```python
#Se sustituyeron los valores nulos de la columna "alimentos_preparados_rotiseria" por valores no nulos hacia atrás con el filtro por columna "backward fill"
data1["alimentos_preparados_rotiseria"] =data1["alimentos_preparados_rotiseria"].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos en todo el dataframe
valores_nulos=data1.isnull().sum().sum()
valores_nulos
```
